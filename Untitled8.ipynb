```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
import matplotlib.pyplot as plt

# تعریف دیتاست سفارشی
class ProteinDataset(Dataset):
    def __init__(self, sequences1, sequences2, labels):
        self.sequences1 = sequences1
        self.sequences2 = sequences2
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.sequences1[idx], self.sequences2[idx], self.labels[idx]

# تعریف شبکه سیامی
class SiameseLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(SiameseLSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

    def forward(self, x):
        _, (hidden, _) = self.lstm(x)
        return hidden[-1]  # بردار ویژگی آخرین لایه پنهان

class SiameseNetworkWithConcat(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(SiameseNetworkWithConcat, self).__init__()
        self.lstm_branch = SiameseLSTM(input_size, hidden_size, num_layers)
        self.fc = nn.Sequential(
            nn.Linear(hidden_size * 2, 128),  # 2 برابر hidden_size به دلیل الحاق
            nn.ReLU(),
            nn.Linear(128, 1)  # پیش‌بینی یک مقدار فاصله
        )

    def forward(self, input1, input2):
        output1 = self.lstm_branch(input1)
        output2 = self.lstm_branch(input2)

        # الحاق خروجی‌ها
        concatenated = torch.cat((output1, output2), dim=1)

        # پیش‌بینی فاصله یا شباهت
        distance = self.fc(concatenated)
        return distance

# تنظیمات مدل
input_size = 128      # تعداد ویژگی‌های ورودی
sequence_length = 100 # طول توالی
hidden_size = 64      # تعداد نرون‌های لایه پنهان
num_layers = 2        # تعداد لایه‌های LSTM
batch_size = 16       # اندازه دسته‌ها
num_epochs = 6        # تعداد تکرارها
learning_rate = 0.001 # نرخ یادگیری
threshold = 0.5       # آستانه برای طبقه‌بندی

# داده‌های مصنوعی
num_samples = 500  # تعداد نمونه‌ها
sequences1 = torch.randn(num_samples, sequence_length, input_size)  # توالی‌های اول
sequences2 = torch.randn(num_samples, sequence_length, input_size)  # توالی‌های دوم
labels = (torch.rand(num_samples) > 0.5).float()  # برچسب‌ها (0 یا 1)

# ساخت Dataset
dataset = ProteinDataset(sequences1, sequences2, labels)

# تقسیم داده‌ها به Train و Test
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# تعریف مدل، تابع زیان و بهینه‌ساز
model = SiameseNetworkWithConcat(input_size, hidden_size, num_layers)
criterion = nn.BCEWithLogitsLoss()  # تابع زیان مناسب برای logits
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# حلقه آموزش
accuracies = []
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    correct_predictions = 0
    total_samples = 0

    for batch_idx, (seq1, seq2, labels) in enumerate(train_loader):
        # پیش‌بینی
        outputs = model(seq1, seq2).squeeze()
        predictions = (torch.sigmoid(outputs) > threshold).float()  # اعمال سیگموید

        # محاسبه loss
        loss = criterion(outputs, labels)
        epoch_loss += loss.item()

        # محاسبه تعداد پیش‌بینی‌های صحیح
        correct_predictions += (predictions == labels).sum().item()
        total_samples += labels.size(0)

        # به‌روزرسانی وزن‌ها
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # محاسبه و ذخیره نرخ صحت
    accuracy = correct_predictions / total_samples
    accuracies.append(accuracy)
    print(f"Epoch [{epoch+1}/{num_epochs}] Completed, Average Loss: {epoch_loss / len(train_loader):.4f}, Accuracy: {accuracy:.4f}")

# آزمون مدل
model.eval()  # حالت ارزیابی
test_correct = 0
test_total = 0

with torch.no_grad():  # غیرفعال کردن گرادیان برای افزایش سرعت
    for seq1, seq2, labels in test_loader:
        # پیش‌بینی
        outputs = model(seq1, seq2).squeeze()
        predictions = (torch.sigmoid(outputs) > threshold).float()  # اعمال سیگموید

        # محاسبه تعداد پیش‌بینی‌های صحیح
        test_correct += (predictions == labels).sum().item()
        test_total += labels.size(0)

# محاسبه دقت داده‌های آزمون
test_accuracy = test_correct / test_total
print(f"Test Accuracy: {test_accuracy:.4f}")

# رسم نمودار دقت
plt.plot(range(1, num_epochs + 1), accuracies, marker='o', linestyle='-', color='b')
plt.title('Training Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
