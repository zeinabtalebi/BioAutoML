```python
# وارد کردن کتابخانه‌های مورد نیاز
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score

# تعریف کلاس Dataset برای مدیریت داده‌ها
class HeadTrajectoryDataset(Dataset):
    """
    کلاس برای مدیریت داده‌های ورودی (مسیر حرکت سر) و برچسب‌ها (مختصات هدف).
    """
    def __init__(self, data, labels):
        self.data = data  # داده‌های مسیر حرکت سر
        self.labels = labels  # برچسب‌های خروجی (مختصات هدف)

    def __len__(self):
        return len(self.data)  # تعداد نمونه‌ها

    def __getitem__(self, idx):
        return torch.tensor(self.data[idx], dtype=torch.float32), torch.tensor(self.labels[idx], dtype=torch.float32)


# تعریف مدل HiBayes-LSTM
class HiBayesLSTM(nn.Module):
    """
    مدل شامل لایه‌های LSTM و لایه‌های بیزی برای پیش‌بینی مسیر حرکت سر.
    """
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(HiBayesLSTM, self).__init__()
        self.hidden_size = hidden_size  # تعداد نرون‌های مخفی
        self.num_layers = num_layers  # تعداد لایه‌های LSTM

        # تعریف لایه LSTM
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

        # لایه‌های بیزی برای میانگین و انحراف معیار
        self.mu = nn.Linear(hidden_size, output_size)  # لایه تخمین میانگین
        self.sigma = nn.Linear(hidden_size, output_size)  # لایه تخمین انحراف معیار

    def forward(self, x):
        # مقداردهی اولیه به حالت‌های مخفی و سلولی
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # عبور داده‌ها از LSTM
        out, _ = self.lstm(x, (h0, c0))
        out = out[:, -1, :]  # استخراج آخرین گام زمانی

        # محاسبه میانگین و انحراف معیار
        mu = self.mu(out)
        sigma = torch.exp(self.sigma(out))  # مثبت کردن انحراف معیار

        # نمونه‌گیری از توزیع گاوسی
        sampled_output = mu + sigma * torch.randn_like(sigma)
        return sampled_output, mu, sigma


# حلقه آموزش مدل
def train_model(model, dataloader, criterion, optimizer, num_epochs=20):
    """
    آموزش مدل با استفاده از داده‌های آموزشی.
    """
    model.train()
    for epoch in range(num_epochs):
        total_loss = 0.0
        for inputs, targets in dataloader:
            inputs, targets = inputs.to(device), targets.to(device)

            # عبور از مدل و محاسبه خطا
            outputs, mu, sigma = model(inputs)
            loss = criterion(mu, targets)

            # به‌روزرسانی وزن‌ها
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
        print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {total_loss / len(dataloader):.4f}")


# حلقه ارزیابی مدل
def evaluate_model_with_metrics(model, dataloader):
    """
    ارزیابی مدل و محاسبه معیارهای MAE و R².
    """
    model.eval()
    predictions = []
    ground_truths = []
    with torch.no_grad():
        for inputs, targets in dataloader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs, _, _ = model(inputs)
            predictions.append(outputs.cpu().numpy())
            ground_truths.append(targets.cpu().numpy())

    # ترکیب نتایج
    predictions = np.concatenate(predictions)
    ground_truths = np.concatenate(ground_truths)

    # محاسبه MAE و R²
    mae = mean_absolute_error(ground_truths, predictions)
    r2 = r2_score(ground_truths, predictions)
    return mae, r2, predictions, ground_truths


# تولید داده‌های مسیر حرکت سر
np.random.seed(42)

# 1000 نمونه، 10 گام زمانی، 3 مختصات (x, y, z)
data = np.cumsum(np.random.randn(1000, 10, 3), axis=1)  # مسیر تجمعی برای حرکت طبیعی‌تر

# مختصات هدف (نقطه نهایی مسیر سر)
labels = data[:, -1, :]  # مختصات گام آخر برای هر نمونه

# ساخت Dataset و DataLoader
dataset = HeadTrajectoryDataset(data, labels)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# تنظیم مدل
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = HiBayesLSTM(input_size=3, hidden_size=64, output_size=3).to(device)

# تعریف معیار خطا و بهینه‌ساز
criterion = nn.MSELoss()  # خطای مربعات میانگین
optimizer = optim.Adam(model.parameters(), lr=0.001)

# آموزش مدل
train_model(model, dataloader, criterion, optimizer, num_epochs=20)

# ارزیابی مدل
mae, r2, predictions, ground_truths = evaluate_model_with_metrics(model, dataloader)

# نمایش نتایج
print(f"Mean Absolute Error (MAE): {mae:.4f}")  # خطای مطلق میانگین
print(f"R² Score: {r2:.4f}")  # ضریب تعیین
print("Sample Predictions:", predictions[:5])  # 5 پیش‌بینی اول
print("Sample Ground Truths:", ground_truths[:5])  # 5 مقدار واقعی اول
```
